```python
pd.crosstab(data["Happiness_Score"],data["Health_Life_Expectancy"])
##counting the occurrences of one column value with respect to another.
```

```python
data.pivot_table(index=["Happiness_Score", "Health_Life_Expectancy"], aggfunc=np.sum)
##aggregate data in a DataFrame.
```

```python
cross_1=pd.crosstab(data["Happiness_Score"],data["Health_Life_Expectancy"])
cross_2=cross_1.stack().reset_index()
cross_2
##reset indexed values.
```

```python
plt.scatter(cross_2["Happiness_Score"],cross_2["Health_Life_Expectancy"],s=np.sqrt(cross_2[0])*1.5,alpha=0.5)
plt.xlim(2.839, 7.587)
plt.ylim(0,1.02525)
plt.xlabel("Happiness_Score")
plt.ylabel("Health_Life_Expectancy")
plt.show()
##visualizations data.
```

```python
data["Health_Life_Expectancy"].plot.line(ylim=[0,1.02525])
plt.xlabel("Happiness_Score")
plt.ylabel("Health_Life_Expectancy")
##visualizations data.
```

```python
scipy.stats.spearmanr(data["Happiness_Score"],data["Health_Life_Expectancy"])
##corrletion value use spearmanr.
```

```python
data["Happiness_Score"].corr(data["Health_Life_Expectancy"],method="pearson")
##corrletion value use pearson.
```
